```python
# Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots look cleaner
sns.set()
```

```python
# Load Dataset

df = pd.read_csv("data.csv")

# Display first 5 rows
df.head()
```

```python
# Check all values
print(df.isna())
```

```python
print(df.isna().sum())
```

```python
# checking duplicate rows
df.duplicated()[df.duplicated()==True]
```

```python
# finding number of rows and columns

print("Shape (rows, columns): ", df.shape,"\n")

print("number of rows: ", df.shape[0])
print("number of columns: ", df.shape[1])
```

```python
# viewing the data types of columns
df.dtypes
```

```python
# Statistical summary
df.describe(include='all')
```

```python
sns.histplot(df[df['price'] < 3000000]['price'], kde=True, color='blue', bins=50)

plt.title('Distribution of House Prices', fontsize=15)
plt.xlabel('Price ($)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
```

```python
filtered_df = df[df['price'] < 3000000]

sns.scatterplot(data=filtered_df, x='sqft_living', y='price', alpha=0.5, color='teal')

# Add labels and title
plt.title('Living Area (sqft) vs. House Price', fontsize=15)
plt.xlabel('Living Area (sqft)', fontsize=12)
plt.ylabel('Price ($)', fontsize=12)
```

```python
filtered_df = df[df['price'] < 3000000]

# Create the Bedrooms plot
plt.figure(figsize=(10, 6))
sns.barplot(data=filtered_df, x='bedrooms', y='price', estimator='mean', errorbar=None)
plt.title('Average Price by Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Average Price ($)')
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=filtered_df, x='bathrooms', y='price', estimator='mean', errorbar=None)
plt.title('Average Price by Number of Bathrooms')
plt.xlabel('Number of Bathrooms')
plt.ylabel('Average Price ($)')
plt.xticks(rotation=45)
```

```python
numerical_df = df.select_dtypes(include=['float64', 'int64'])
corr_matrix = numerical_df.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, linewidths=.5)
plt.title('Correlation Heatmap of Property Features', fontsize=15)
```

```python
yearly_trends = df.groupby('yr_built')['price'].median().reset_index()
plt.figure(figsize=(12, 6))
sns.lineplot(data=yearly_trends, x='yr_built', y='price')
plt.title('Construction Trends: Median House Price by Year Built', fontsize=15)
plt.xlabel('Year Built', fontsize=12)
plt.ylabel('Median Price ($)', fontsize=12)
```
